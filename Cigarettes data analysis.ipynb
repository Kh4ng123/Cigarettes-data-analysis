```python
import pandas as pd
import numpy as np
df = pd.read_csv('framingham.csv')
smokers = df["cigsPerDay"].dropna()
numberOfObservation = len(smokers)
print("This is the number of observations:", numberOfObservation)
print("This is the mean:", round(smokers.mean(),1))
print("This is the max:", round(smokers.max(),1))
print("This is the standard deviation:", round(smokers.std(), 1))
```

```python
sample = np.random.choice(df.cigsPerDay.dropna(), 100)
sample_mean = np.mean(sample)
print("This is the sample mean: ", round(sample_mean, 2))
```

```python
import matplotlib.pyplot as plt
sample_means_list = []
for i in range(50):
    sample = np.random.choice(df.cigsPerDay.dropna(), 100)
    sample_means_list.append(np.mean(sample))

sampling_dist_mean = np.mean(sample_means_list)
true_mean = smokers.mean()

plt.hist(sample_means_list, bins = 20)
plt.title("Distribution of cigarettes consumed") # title
plt.xlabel("Cigarettes") # x axis label
plt.ylabel("Count") # y axis label
plt.axvline(sampling_dist_mean, color = "red", label = "Sample mean")
plt.axvline(true_mean, color = "black", label = "True mean")
plt.legend()
plt.show()
```
